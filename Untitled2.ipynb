```python
url = input('뒷자리 url 적으셈(ex: watch?v=2A6pygV4Z04 ) : ')
url = 'https://www.youtube.com/'+url
num_scroll = int(input('댓글 몇 줄 가져올까 씨발아? (50 recommend) : '))
```

```python
driver = webdriver.Chrome(r'c:\Users\korea\chromedriver')
driver.get(url)
```

```python
reply = driver.find_element_by_tag_name('body')
```

```python
reply_user = []
reply_content = []
reply_likes = []
```

```python
while num_scroll:
    reply.send_keys(Keys.PAGE_DOWN)
    time.sleep(0.1)
    num_scroll -= 1
```

```python
html0 = driver.page_source
html = BeautifulSoup(html0, 'html.parser')
```

```python
driver.close()
```
